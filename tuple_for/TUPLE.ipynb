# 📘 Python Set – Complete Documentation

## 🔹 What is a Set?

A **set** in Python is a collection of **unique** and **unordered** elements.

* No duplicate values allowed
* No indexing (you can’t access elements by position)
* Mutable (you can add/remove items)

---

## 🔹 Creating a Set

```python
# Using curly braces
my_set = {1, 2, 3, 4}

# Using set() function
my_set2 = set([1, 2, 3, 4])

print(my_set)
```

---

## 🔹 Important Note

```python
empty_set = set()   # सही way
wrong_set = {}      # This creates a dictionary, not a set
```

---

## 🔹 Set Properties

| Property    | Description             |
| ----------- | ----------------------- |
| Unordered   | No fixed order          |
| Unique      | No duplicate elements   |
| Mutable     | Can be modified         |
| No indexing | Cannot access via index |

---

## 🔹 Example (Uniqueness)

```python
numbers = {1, 2, 2, 3, 4, 4}
print(numbers)
```

👉 Output:

```
{1, 2, 3, 4}
```

---

## 🔹 Adding Elements

```python
my_set = {1, 2, 3}

my_set.add(4)
print(my_set)
```

---

## 🔹 Removing Elements

```python
my_set.remove(2)   # Error if not found
my_set.discard(5)  # No error if not found
```

---

## 🔹 Set Operations

### 1. Union (Combine sets)

```python
A = {1, 2, 3}
B = {3, 4, 5}

print(A | B)
```

👉 Output:

```
{1, 2, 3, 4, 5}
```

---

### 2. Intersection (Common elements)

```python
print(A & B)
```

👉 Output:

```
{3}
```

---

### 3. Difference (A - B)

```python
print(A - B)
```

👉 Output:

```
{1, 2}
```

---

### 4. Symmetric Difference

```python
print(A ^ B)
```

👉 Output:

```
{1, 2, 4, 5}
```

---

## 🔹 Looping Through a Set

```python
for item in my_set:
    print(item)
```

---

## 🔹 Checking Membership

```python
if 2 in my_set:
    print("Exists")
```

---

## 🔹 Set Methods

| Method         | Description                         |
| -------------- | ----------------------------------- |
| add()          | Add element                         |
| remove()       | Remove element (error if not found) |
| discard()      | Remove element safely               |
| pop()          | Remove random element               |
| clear()        | Remove all elements                 |
| union()        | Combine sets                        |
| intersection() | Common elements                     |
| difference()   | Difference between sets             |

---

## 🔹 Real-Life Example

👉 Remove duplicate values from a list

```python
marks = [90, 85, 90, 70, 85]

unique_marks = set(marks)
print(unique_marks)
```

👉 Output:

```
{90, 85, 70}
```

---

## 🔹 Frozen Set (Immutable Set)

```python
fs = frozenset([1, 2, 3])
# fs.add(4) ❌ Error (cannot modify)
```
